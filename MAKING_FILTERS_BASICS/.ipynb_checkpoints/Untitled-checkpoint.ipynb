```python
import numpy as np
import pandas as pd
import cv2
import matplotlib.pyplot as plt
```

```python
eye_identity = cv2.CascadeClassifier('./frontalEyes35x16.xml')
nose_identity = cv2.CascadeClassifier('./Nose18x15.xml')
```

```python
pic = cv2.imread("./Before.png")
pic=cv2.cvtColor(pic,cv2.COLOR_BGRA2RGBA)

eye = eye_identity.detectMultiScale(pic,1.1,5)
nose = nose_identity.detectMultiScale(pic,1.1,5)

filter_eye = cv2.imread("./glasses.png",cv2.IMREAD_UNCHANGED)
filter_eye = cv2.cvtColor(filter_eye,cv2.COLOR_BGRA2RGBA)
filter_nose = cv2.imread("./mustache.png",cv2.IMREAD_UNCHANGED)
filter_nose = cv2.cvtColor(filter_nose,cv2.COLOR_BGRA2RGBA)
```

```python
eye,nose
```

```python
x,y,w,h = eye[0]
x1,y1,w1,h1 = nose[0]

filter_eye = cv2.resize(filter_eye,(w,h))
filter_nose = cv2.resize(filter_nose,(w1,h1))
```

```python
pic.shape,filter_eye.shape
```

```python
for i in range(filter_eye.shape[0]):
    for j in range(filter_eye.shape[1]):
        if(filter_eye[i,j,3]>0):
            pic[y+i,x+j,:-1]=filter_eye[i,j,:-1]
            
            
for i in range(filter_nose.shape[0]):
    for j in range(filter_nose.shape[1]):
        if(filter_nose[i,j,3]>0):
            pic[y1+i,x1+j,:-1]=filter_nose[i,j,:-1]            
# pic[y:y+h,x:x+w,:] = filter_eye[:,:,:]
```

```python
cap = cv2.VideoCapture(0)
face_identify = cv2.CascadeClassifier("./haarcascade_frontalface_alt.xml")
eye_identify = cv2.CascadeClassifier("./frontalEyes35x16.xml")
nose_identify = cv2.CascadeClassifier("./Nose18x15.xml")
while True:
    
    ret,frame = cap.read()
    if(ret == False):
        continue
      
    gray = cv2.cvtColor(frame,cv2.COLOR_BGR2RGB)
    faces = face_identify.detectMultiScale(gray,1.3,5)
    eyes = eye_identify.detectMultiScale(gray,1.1,5)
    nose = nose_identify.detectMultiScale(gray,1.1,5)
       
    for (x,y,w,h) in faces:
        cv2.rectangle(frame,(x,y),(x+w,y+h),(255,0,0),2)
        
        
        for (x1,y1,w1,h1) in eyes:
            #cv2.rectangle(frame,(x1,y1),(x1+w1,y1+h1),(0,225,0),2)
        
            overlay=cv2.imread("glasses.png",cv2.IMREAD_UNCHANGED)
            overlay=cv2.cvtColor(overlay,cv2.COLOR_BGRA2RGBA)

            overlay=cv2.resize(overlay,(w1,h1))
            #print(frame.shape,overlay.shape)
            for i in range(overlay.shape[0]):
                for j in range(overlay.shape[1]):
                    if(overlay[i,j,3]>0):
                        frame[y1+i,x1+j,:]=overlay[i,j,:-1]
        
        
        
        
    cv2.imshow("Video",frame)    
    
        
    key = cv2.waitKey(1) & 0xFF
    if(key == ord('q')):
        break;
        
cap.release()
cv2.destroyAllWindows()
        
```
